```python
# pc 2/20/2026
import pandas as pd
from simpledbf import Dbf5

# -----------------------------
# LOAD BG DBF PROPERLY
# -----------------------------
dbf = Dbf5("acs2024_bg_with_gq_allocated_scaled.dbf")
bg = dbf.to_dataframe()
bg.columns = bg.columns.str.strip()

# -----------------------------
# LOAD TAZ CSV
# -----------------------------
taz = pd.read_csv("TAZ_allocated_clean.csv")

# -----------------------------
# COMPUTE TOTALS
# -----------------------------
bg_totals = {
    "TOT_HH": bg["TOT_HH"].sum(),
    "HH_POP": bg["HH_POP"].sum(),
    "TOT_POP": bg["TOT_POP"].sum(),
    "EMP_RES": bg["EMP_RES"].sum()
}

taz_totals = {
    "TOT_HH": taz["TOT_HH_alloc"].sum(),
    "HH_POP": taz["HH_POP_alloc"].sum(),
    "TOT_POP": taz["TOT_POP_alloc"].sum(),
    "EMP_RES": taz["EMP_RES_alloc"].sum()
}

print("\nValidation Summary\n")

for key in bg_totals:
    diff = taz_totals[key] - bg_totals[key]
    pct = diff / bg_totals[key] * 100
    print(f"{key}:")
    print(f"  BG Total  = {bg_totals[key]:,.0f}")
    print(f"  TAZ Total = {taz_totals[key]:,.0f}")
    print(f"  Diff      = {diff:,.6f} ({pct:.6f}%)\n")
```
